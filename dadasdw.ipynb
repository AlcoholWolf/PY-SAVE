```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib
%matplotlib inline

import pandas_profiling as pp

import warnings
warnings.filterwarnings(action='ignore')

plt.style.use('dark_background')
```

```python
train = pd.read_csv('/content/train.csv')
test = pd.read_csv('/content/test.csv')

train= train.drop(['index'],axis=1)

# train.info()
# train.describe(include='all')
```

```python
numerical_columns = train.select_dtypes(exclude='object').columns.tolist()
numerical_columns.remove('quality')
def show_dist_plot(df, columns):
    for column in columns:
        f, ax = plt.subplots(figsize=(16,4))
        sns.stripplot(x=df['quality'],y=df[column],hue=df['quality'],size=2)
        
show_dist_plot(train, numerical_columns)
```
